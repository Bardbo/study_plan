Tensors（张量）与Numpy中的数组ndarrays十分相似，只不过tensors可以在GPU上运行

```python
import torch
import numpy as np
```

## 1.初始化一个Tensor

```python
# 从数据创建
data = [[1, 2],[3, 4]]
x_data = torch.tensor(data)

# 从numpy数组创建
np_array = np.array(data)
x_np = torch.from_numpy(np_array)

# 从其它的tensor创建
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")

# 随机或根据常量创建
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

## 2.tensor的属性

```python
tensor = torch.rand(3,4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

## 3.操作tensor

```python
# 将tensor转移到GPU中
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# tensor的操作和numpy差不多
tensor = torch.ones(4, 4)
print('First row: ',tensor[0])
print('First column: ', tensor[:, 0])
print('Last column:', tensor[..., -1])
tensor[:,1] = 0
print(tensor)
```

```python
# 合并的操作
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
# 数学运算
# 下面的y1,y2,y3都会得到一样的结果（矩阵乘法）
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(tensor)
torch.matmul(tensor, tensor.T, out=y3)


# 逐元素的乘法. z1, z2, z3 will have the same value
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

```python
# 将仅含有一个元素的tensor转为python中的标量
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```

```python
# In-place 操作
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

In-place虽然会节省一些内存，但可能因为疏忽而出现错误，因此不建议使用

## 4.连接Numpy

```python
# tensor转numpy
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
# 改变tensor也会改变这个numpy数组
# ！！！注意这个
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")

# 改变数组也会改变tensor
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
