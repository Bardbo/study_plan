## 1.热身：Numpy的一个简单拟合例子

```python
# 导库
import numpy as np
import math
```

```python
# 编一个sin数据
# 等间距的在-Π到Π之间生成2000个x值
x = np.linspace(-math.pi, math.pi, 2000)
y = np.sin(x)
print(x.shape, y.shape)
```

```python
# 随机初始化权重，后面设置的拟合曲线为 y = a + b x + c x^2 + d x^3
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
d = np.random.randn()
print(a, b, c, d)
```

```python
# 学习率
learning_rate = 1e-6
# 迭代2000次
for t in range(2000):
    # 前向传播，计算预测的y值
    y_pred = a + b * x + c * x ** 2 + d * x ** 3
    
    # 计算损失（误差），并每隔100次打印
    loss = np.square(y_pred - y).sum()
    if t % 100 == 99:
        print(t, loss)
    
    # 反向传播，计算梯度
    grad_y_pred = 2.0 * (y_pred - y)
    grad_a = grad_y_pred.sum()
    grad_b = (grad_y_pred * x).sum()
    grad_c = (grad_y_pred * x ** 2).sum()
    grad_d = (grad_y_pred * x ** 3).sum()
    
    # 更新权重
    # -= 的写法等价于 a = a - learning_rate * grad_a
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    c -= learning_rate * grad_c
    d -= learning_rate * grad_d
# 打印出拟合曲线公式
# f 的写法等价于 "Result: y = {0} + {1} x + {2} x^2 + {3} x^3".format(a, b, c, d)
print(f'Result: y = {a} + {b} x + {c} x^2 + {d} x^3')
```

```python
# 导入绘图库
import matplotlib.pyplot as plt

# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题

# 绘制原数据的散点图
plt.scatter(x, y)
# 绘制拟合曲线
plt.plot(x, a + b * x + c * x ** 2 + d * x ** 3, c='r')
# 添加图例
plt.legend(['拟合曲线', '原数据'])
```

## 2.pytorch完成上述过程

```python
# 导库
import torch
import math
```

```python
dtype = torch.float
# device = torch.device("cpu") # 没有GPU就取消此行注释，使用CPU
device = torch.device("cuda:0") # 使用GPU

# 还是使用前面一样的数据，这次使用torch来生成
# 注意指定了是使用CPU还是GPU，以及数据类型
x = torch.linspace(-math.pi, math.pi, 2000, device=device, dtype=dtype)
y = torch.sin(x)

# 随机初始化权重，和上面一样指定了GPU和类型
a = torch.randn((), device=device, dtype=dtype)
b = torch.randn((), device=device, dtype=dtype)
c = torch.randn((), device=device, dtype=dtype)
d = torch.randn((), device=device, dtype=dtype)

# 同样的步骤，不进行再次解释
learning_rate = 1e-6
for t in range(2000):
    # Forward pass: compute predicted y
    y_pred = a + b * x + c * x ** 2 + d * x ** 3

    # Compute and print loss
    loss = (y_pred - y).pow(2).sum().item()
    if t % 100 == 99:
        print(t, loss)

    # Backprop to compute gradients of a, b, c, d with respect to loss
    grad_y_pred = 2.0 * (y_pred - y)
    grad_a = grad_y_pred.sum()
    grad_b = (grad_y_pred * x).sum()
    grad_c = (grad_y_pred * x ** 2).sum()
    grad_d = (grad_y_pred * x ** 3).sum()

    # Update weights using gradient descent
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    c -= learning_rate * grad_c
    d -= learning_rate * grad_d


print(f'Result: y = {a.item()} + {b.item()} x + {c.item()} x^2 + {d.item()} x^3')
```

```python
# 再次绘图观察

# 数据从GPU转到CPU
x = x.cpu()
y = y.cpu()
a, b, c, d = a.cpu(), b.cpu(), c.cpu(), d.cpu()

# 绘制原数据的散点图
plt.scatter(x, y)
# 绘制拟合曲线
plt.plot(x, a + b * x + c * x ** 2 + d * x ** 3, c='r')
# 添加图例
plt.legend(['拟合曲线', '原数据'])
```

## 3.pytorch中更好的方式完成上面的例子

```python
import torch
import math

dtype = torch.float
# device = torch.device("cpu")
device = torch.device("cuda:0")  # Uncomment this to run on GPU

x = torch.linspace(-math.pi, math.pi, 2000, device=device, dtype=dtype)
y = torch.sin(x)

# 注意此时权重的初始化中多了requires_grad=True
# 这表示我们将要在反向传播时计算这些tensor的梯度
a = torch.randn((), device=device, dtype=dtype, requires_grad=True)
b = torch.randn((), device=device, dtype=dtype, requires_grad=True)
c = torch.randn((), device=device, dtype=dtype, requires_grad=True)
d = torch.randn((), device=device, dtype=dtype, requires_grad=True)

learning_rate = 1e-6
for t in range(2000):
    y_pred = a + b * x + c * x ** 2 + d * x ** 3
    
    # 注意这里的loss和前面的loss的不同之处
    # 此时的loss是shape为（1，）的tensor
    # 前面的loss中使用了.item, 相当于一个标量
    loss = (y_pred - y).pow(2).sum()
    if t % 100 == 99:
        print(t, loss.item())
        
    # 这里就是pytorch中更好的方式！
    # 即梯度的自动计算，反向传播 autograd
    # 设置了requires_grad=True的tensor均会自动计算梯度
    loss.backward()
    
    # 这里表示在权重更新时不使用梯度的自动计算
    with torch.no_grad():
        a -= learning_rate * a.grad
        b -= learning_rate * b.grad
        c -= learning_rate * c.grad
        d -= learning_rate * d.grad

        # 在权重更新后将梯度清零
        a.grad = None
        b.grad = None
        c.grad = None
        d.grad = None

print(f'Result: y = {a.item()} + {b.item()} x + {c.item()} x^2 + {d.item()} x^3') 
```

```python
# 再次绘图观察

# 数据从GPU转到CPU
x = x.cpu()
y = y.cpu()
a, b, c, d = a.detach().cpu(), b.detach().cpu(), c.detach().cpu(), d.detach().cpu()

# 绘制原数据的散点图
plt.scatter(x, y)
# 绘制拟合曲线
plt.plot(x, a + b * x + c * x ** 2 + d * x ** 3, c='r')
# 添加图例
plt.legend(['拟合曲线', '原数据'])
```

## 4.自定义autograd

仍是使用上面的例子，但是这次我们使用的拟合曲线为$y=a+bP_{3}(c+dx)$，其中$P_{3}(x)=\frac{1}{2}(5x^3-3x)$

这里$P_{3}(x)$关于$x$的导数为$1.5(5x^2-1)$

在pytorch中自定义autograd通过继承torch.autograd.Function类实现，需要自己写forward和backward方法，也即前向和后向

```python
import torch
import math

# 子类，自定义P_3的autograd
class LegendrePolynomial3(torch.autograd.Function):
    """
    We can implement our own custom autograd Functions by subclassing
    torch.autograd.Function and implementing the forward and backward passes
    which operate on Tensors.
    """
    
    # 静态方法，可以不实例化使用，也可以实例化使用
    @staticmethod
    def forward(ctx, input):
        """
        In the forward pass we receive a Tensor containing the input and return
        a Tensor containing the output. ctx is a context object that can be used
        to stash information for backward computation. You can cache arbitrary
        objects for use in the backward pass using the ctx.save_for_backward method.
        """
        # 为反向传播存放数据信息，下面的backward方法中可以取出来
        ctx.save_for_backward(input)
        return 0.5 * (5 * input ** 3 - 3 * input)

    @staticmethod
    def backward(ctx, grad_output):
        """
        In the backward pass we receive a Tensor containing the gradient of the loss
        with respect to the output, and we need to compute the gradient of the loss
        with respect to the input.
        """
        input, = ctx.saved_tensors
        return grad_output * 1.5 * (5 * input ** 2 - 1)

dtype = torch.float
# device = torch.device("cpu")
device = torch.device("cuda:0")  # Uncomment this to run on GPU

x = torch.linspace(-math.pi, math.pi, 2000, device=device, dtype=dtype)
y = torch.sin(x)

# torch.full 这里不是随机初始化，而是指定了各自的数值，与合适的拟合值差的不多
a = torch.full((), 0.0, device=device, dtype=dtype, requires_grad=True)
b = torch.full((), -1.0, device=device, dtype=dtype, requires_grad=True)
c = torch.full((), 0.0, device=device, dtype=dtype, requires_grad=True)
d = torch.full((), 0.3, device=device, dtype=dtype, requires_grad=True)

print(a, b, c, d)
```

```python
learning_rate = 5e-6
for t in range(2000):
    # To apply our Function, we use Function.apply method. We alias this as 'P3'.
    # 应用我们自定义的autograd
    P3 = LegendrePolynomial3.apply

    # Forward pass: compute predicted y using operations; we compute
    # P3 using our custom autograd operation.
    y_pred = a + b * P3(c + d * x)

    # Compute and print loss
    loss = (y_pred - y).pow(2).sum()
    if t % 100 == 99:
        print(t, loss.item())

    # Use autograd to compute the backward pass.
    loss.backward()

    # Update weights using gradient descent
    with torch.no_grad():
        a -= learning_rate * a.grad
        b -= learning_rate * b.grad
        c -= learning_rate * c.grad
        d -= learning_rate * d.grad

        # Manually zero the gradients after updating weights
        a.grad = None
        b.grad = None
        c.grad = None
        d.grad = None

print(f'Result: y = {a.item()} + {b.item()} * P3({c.item()} + {d.item()} x)')
```

```python
# 再次绘图观察

# 数据从GPU转到CPU
x = x.cpu()
y = y.cpu()
# 实际上这里我们也可以使用.item()的方法取出标量
a, b, c, d = a.item(), b.item(), c.item(), d.item()

# 绘制原数据的散点图
plt.scatter(x, y)
# 绘制拟合曲线
plt.plot(x, a + b * P3(c + d * x), c='r')
# 添加图例
plt.legend(['拟合曲线', '原数据'])
```

## 5.pytorch中的nn模块

这里就是pytorch中很重要的一个模块了，包含神经网络中的各种layers（层），还有损失函数等等

下面我们将原始的多项式拟合曲线用nn模块的方式再实现一下

```python
import torch
import math

x = torch.linspace(-math.pi, math.pi, 2000)
y = torch.sin(x)

# y = a + b x + c x^2 + d x^3
# 实际上这种多项式回归和线性回归是一样的，我们把特征看成 x， x^2 ，x^3即可
p = torch.tensor([1, 2, 3])
# x.unsqueeze(-1)的shape是（2000，1），相当于增加维度
# xx的shape是（2000，3），因为p的维度是（3，），通过pow运算的广播机制就变成了（2000，3）
# 即第一列是x，第二列是x^2，第三列是x^3
xx = x.unsqueeze(-1).pow(p)

# 这里用到了nn中的3个方法Sequential、Linear、Flatten，来构建拟合模型
# 后两个是layers，前一个的作用就是串联输入的层
# Linear就是线性模型，如 a + b x + c x^2 + d x^3，此时我们输入的维度是3，输出的维度是1
# Flatten就是展开，这里是将（2000，1）的Linear层输出展平为（2000，）
# 这样就可以与y的shape对应起来，进行损失的计算
model = torch.nn.Sequential(
    torch.nn.Linear(3, 1),
    torch.nn.Flatten(0, 1)
)

# nn模块中同时还包含了主流的损失函数，这里我们使用均方误差损失（常用的回归损失）
# 并和之前一样会对损失进行求和
loss_fn = torch.nn.MSELoss(reduction='sum')

learning_rate = 1e-6
for t in range(2000):

    # Forward pass: compute predicted y by passing x to the model. Module objects
    # override the __call__ operator so you can call them like functions. When
    # doing so you pass a Tensor of input data to the Module and it produces
    # a Tensor of output data.
    # 将输入传入模型获取输出
    y_pred = model(xx)

    # Compute and print loss. We pass Tensors containing the predicted and true
    # values of y, and the loss function returns a Tensor containing the
    # loss.
    # 计算损失
    loss = loss_fn(y_pred, y)
    if t % 100 == 99:
        print(t, loss.item())

    # Zero the gradients before running the backward pass.
    # 在反向传播之前进行梯度清零
    model.zero_grad()

    # Backward pass: compute gradient of the loss with respect to all the learnable
    # parameters of the model. Internally, the parameters of each Module are stored
    # in Tensors with requires_grad=True, so this call will compute gradients for
    # all learnable parameters in the model.
    # 误差的反向传播
    loss.backward()

    # Update the weights using gradient descent. Each parameter is a Tensor, so
    # we can access its gradients like we did before.
    # 参数更新（权重更新）
    with torch.no_grad():
        for param in model.parameters():
            param -= learning_rate * param.grad

# You can access the first layer of `model` like accessing the first item of a list
# 我们可以从model中取出我们想要的层
# 这里我们的模型实际上是一个线性模型，我们取出线性层
# 便于后面取出线性层的参数，也就是a，b，c，d
linear_layer = model[0]

# For linear layer, its parameters are stored as `weight` and `bias`.
# a就是bias，其余权重按顺序取即可
print(f'Result: y = {linear_layer.bias.item()} + {linear_layer.weight[:, 0].item()} x + '
      + f'{linear_layer.weight[:, 1].item()} x^2 + {linear_layer.weight[:, 2].item()} x^3')
```

```python
# 再次绘图观察

# 数据从GPU转到CPU
x = x.cpu()
y = y.cpu()
# 权重值
a, b = linear_layer.bias.item(), linear_layer.weight[:, 0].item()
c, d = linear_layer.weight[:, 1].item(), linear_layer.weight[:, 2].item()

# 绘制原数据的散点图
plt.scatter(x, y)
# 绘制拟合曲线
plt.plot(x, a + b * x + c * x ** 2 + d * x ** 3, c='r')
# 添加图例
plt.legend(['拟合曲线', '原数据'])
```

## 6.pytorch中的optim

上面我们对于权重的更新都是手动的使用torch.no_grad()的方法，这对于SGD（随机梯度下降）而言可能还比较方便

但如果我们优化器更加复杂，比如使用AdaGrad, RMSProp, Adam这些方法时，再手动的实现就比较复杂了

在pytorch的optim模块中就已经实现了一些流行的复杂优化器

还是上面这个例子，我们接下来不仅使用nn，还会使用optim中的RMSprop算法来进行优化

```python
import torch
import math

x = torch.linspace(-math.pi, math.pi, 2000)
y = torch.sin(x)

p = torch.tensor([1, 2, 3])
xx = x.unsqueeze(-1).pow(p)

model = torch.nn.Sequential(
    torch.nn.Linear(3, 1),
    torch.nn.Flatten(0, 1)
)

loss_fn = torch.nn.MSELoss(reduction='sum')

learning_rate = 1e-3
# 这里设置好所使用的优化器，它应该传入需要优化的参数，即model.parameters()
# 并指定学习率
optimizer = torch.optim.RMSprop(model.parameters(), lr=learning_rate)

for t in range(2000):
    # Forward pass: compute predicted y by passing x to the model.
    y_pred = model(xx)

    # Compute and print loss.
    loss = loss_fn(y_pred, y)
    if t % 100 == 99:
        print(t, loss.item())

    # Before the backward pass, use the optimizer object to zero all of the
    # gradients for the variables it will update (which are the learnable
    # weights of the model). This is because by default, gradients are
    # accumulated in buffers( i.e, not overwritten) whenever .backward()
    # is called. Checkout docs of torch.autograd.backward for more details.
    optimizer.zero_grad()

    # Backward pass: compute gradient of the loss with respect to model
    # parameters
    loss.backward()

    # Calling the step function on an Optimizer makes an update to its
    # parameters
    # 此时的参数更新就不需要手动编写了，调用optimizer的step方法即可
    optimizer.step()

linear_layer = model[0]

print(f'Result: y = {linear_layer.bias.item()} + {linear_layer.weight[:, 0].item()} x + '
      + f'{linear_layer.weight[:, 1].item()} x^2 + {linear_layer.weight[:, 2].item()} x^3')
```

```python
# 再次绘图观察

# 数据从GPU转到CPU
x = x.cpu()
y = y.cpu()
# 权重值
a, b = linear_layer.bias.item(), linear_layer.weight[:, 0].item()
c, d = linear_layer.weight[:, 1].item(), linear_layer.weight[:, 2].item()

# 绘制原数据的散点图
plt.scatter(x, y)
# 绘制拟合曲线
plt.plot(x, a + b * x + c * x ** 2 + d * x ** 3, c='r')
# 添加图例
plt.legend(['拟合曲线', '原数据'])
```

## 7.自定义编写nn模块

nn模块中提供很多层，但有时候仍不满足我们的需求，我们也可以通过继承torch.nn.Module来实现自定义的层

接下来我们就定义一个多项式层（之前我们是通过pow的幂次变换将多项式转化为了等价的线性模型）

```python
import torch
import math

class Polynomial3(torch.nn.Module):
    def __init__(self):
        """
        In the constructor we instantiate four parameters and assign them as
        member parameters.
        """
        super().__init__()
        # 创建我们所需要的四个参数a，b，c，d
        self.a = torch.nn.Parameter(torch.randn(()))
        self.b = torch.nn.Parameter(torch.randn(()))
        self.c = torch.nn.Parameter(torch.randn(()))
        self.d = torch.nn.Parameter(torch.randn(()))
    
    # 编写我们的计算形式
    def forward(self, x):
        """
        In the forward function we accept a Tensor of input data and we must return
        a Tensor of output data. We can use Modules defined in the constructor as
        well as arbitrary operators on Tensors.
        """
        return self.a + self.b * x + self.c * x ** 2 + self.d * x ** 3
    
    # 和python中的类一样，我们也可以添加一些其他方法
    # 这个方法就是返回拟合曲线的公式（字符串）
    def string(self):
        """
        Just like any class in Python, you can also define custom method on PyTorch modules
        """
        return f'y = {self.a.item()} + {self.b.item()} x + {self.c.item()} x^2 + {self.d.item()} x^3'
```

```python
x = torch.linspace(-math.pi, math.pi, 2000)
y = torch.sin(x)

# 这里我们就使用了自定义的多项式层作为我们的模型
model = Polynomial3()

# 损失函数使用MSELoss
criterion = torch.nn.MSELoss(reduction='sum')
# 优化器使用SGD
optimizer = torch.optim.SGD(model.parameters(), lr=1e-6)
for t in range(2000):
    # Forward pass: Compute predicted y by passing x to the model
    y_pred = model(x)

    # Compute and print loss
    loss = criterion(y_pred, y)
    if t % 100 == 99:
        print(t, loss.item())

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
# 这里我们调用了我们自定义的string方法来输出模型拟合曲线的公式结果
print(f'Result: {model.string()}')
```

```python
# 再次绘图观察

# 数据从GPU转到CPU
x = x.cpu()
y = y.cpu()
# 权重值
a = -0.013710835948586464
b = 0.8248897790908813
c = 0.002365347696468234
d = -0.088799849152565

# 绘制原数据的散点图
plt.scatter(x, y)
# 绘制拟合曲线
plt.plot(x, a + b * x + c * x ** 2 + d * x ** 3, c='r')
# 添加图例
plt.legend(['拟合曲线', '原数据'])
```

## 8.pytorch的控制流实现权重共享

为了使用控制流（for循环）来实现权重共享，接下来将实现一个比较奇怪的模型，这个模型出了拥有a，b，c，d的四个参数外

还增加一个参数e，但是与之前的模型不同的是，e对应的x的阶数在每一次传播是随机的

比如可能是单独的4阶， $y = a + b x + c x^2 + d x^3 + e x^4$

也有可能是一个4阶和一个5阶,但权重共享， $y = a + b x + c x^2 + d x^3 + e x^4 + e x^5$

也有可能只有原来的3阶

```python
import random
import torch
import math

# 这里就自定义了这个奇奇怪怪的模型
class DynamicNet(torch.nn.Module):
    def __init__(self):
        """
        In the constructor we instantiate five parameters and assign them as members.
        """
        super().__init__()
        self.a = torch.nn.Parameter(torch.randn(()))
        self.b = torch.nn.Parameter(torch.randn(()))
        self.c = torch.nn.Parameter(torch.randn(()))
        self.d = torch.nn.Parameter(torch.randn(()))
        self.e = torch.nn.Parameter(torch.randn(()))

    def forward(self, x):
        """
        For the forward pass of the model, we randomly choose either 4, 5
        and reuse the e parameter to compute the contribution of these orders.

        Since each forward pass builds a dynamic computation graph, we can use normal
        Python control-flow operators like loops or conditional statements when
        defining the forward pass of the model.

        Here we also see that it is perfectly safe to reuse the same parameter many
        times when defining a computational graph.
        """
        y = self.a + self.b * x + self.c * x ** 2 + self.d * x ** 3
        # 通过for循环和随机数使得模型在传播时是随机的，有可能使用参数e的4阶等等
        for exp in range(4, random.randint(4, 6)):
            y = y + self.e * x ** exp
        return y

    def string(self):
        """
        Just like any class in Python, you can also define custom method on PyTorch modules
        """
        return f'y = {self.a.item()} + {self.b.item()} x + {self.c.item()} x^2 + {self.d.item()} x^3 + {self.e.item()} x^4 ? + {self.e.item()} x^5 ?'


# Create Tensors to hold input and outputs.
x = torch.linspace(-math.pi, math.pi, 2000)
y = torch.sin(x)

# Construct our model by instantiating the class defined above
model = DynamicNet()

# Construct our loss function and an Optimizer. Training this strange model with
# vanilla stochastic gradient descent is tough, so we use momentum
criterion = torch.nn.MSELoss(reduction='sum')

# 注意这里传入了动量参数
optimizer = torch.optim.SGD(model.parameters(), lr=1e-8, momentum=0.9)

# 30000次迭代
for t in range(30000):
    # Forward pass: Compute predicted y by passing x to the model
    y_pred = model(x)

    # Compute and print loss
    loss = criterion(y_pred, y)
    if t % 2000 == 1999:
        print(t, loss.item())

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

print(f'Result: {model.string()}')
```
